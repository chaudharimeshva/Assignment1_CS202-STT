```python
import pandas as pd

# Load your Lab 3 dataset
df = pd.read_csv("/content/Lab3_Final.csv")

# General counts
print("Total commits:", df["Hash"].nunique())
print("Total files:", df["Filename"].nunique())
print("Average files per commit:", df.groupby("Hash")["Filename"].count().mean())

# Structural metrics
print("Average MI Change:", df["MI_Change"].mean())
print("Average CC Change:", df["CC_Change"].mean())
print("Average LOC Change:", df["LOC_Change"].mean())

# Similarity metrics
print("Average Semantic Similarity:", df["Semantic_Similarity"].mean())
print("Average Token Similarity:", df["Token_Similarity"].mean())

# Classification distributions
print("Semantic Classification:", df["Semantic_Class"].value_counts().to_dict())
print("Token Classification:", df["Token_Class"].value_counts().to_dict())
print("Agreement Distribution:", df["Classes_Agree"].value_counts().to_dict())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Lab3_Final.csv")

# Calculate averages
metrics_avg = {
    "MI Change": df["MI_Change"].mean(),
    "CC Change": df["CC_Change"].mean(),
    "LOC Change": df["LOC_Change"].mean(),
    # "Semantic Similarity": df["Semantic_Similarity"].mean(),
    # "Token Similarity": df["Token_Similarity"].mean()
}

# Plot bar chart
plt.figure(figsize=(8,5))
plt.bar(metrics_avg.keys(), metrics_avg.values())
plt.title("Average Values of Metrics in Bug-Fix Commits")
plt.ylabel("Average Value")
plt.xticks(rotation=30)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
